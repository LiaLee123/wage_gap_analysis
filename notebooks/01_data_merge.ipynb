```python
salary_df['CODGEO'] = salary_df['CODGEO'].astype(str).str.zfill(5)

salary_clean = salary_df[['CODGEO', 'LIBGEO', 'SNHM19', 'SNHMF19', 'SNHMH19']].copy()


eps = 1e-9
salary_clean['log_gender_gap'] = (
    np.log(salary_clean['SNHMH19'] + eps) - np.log(salary_clean['SNHMF19'] + eps)
)

print(salary_clean.head())

salary_clean.to_csv("../data/processed/salary_clean.csv", index=False)
```

```python
pop_df['CODGEO'] = pop_df['CODGEO'].astype(str).str.zfill(5)


pop_clean = pop_df[['CODGEO', 'P19_POP', 'P19_POP3044', 'P19_POPF', 'C19_POP15P_CS3']].copy()


salary_with_pop = salary_clean.merge(pop_clean, on='CODGEO', how='left')

print(salary_with_pop.head())


```

```python
import unicodedata 
import re 

def clean_city_name(name):


    name = name.lower().strip()

    name = ''.join(
        c for c in unicodedata.normalize('NFD', name)
        if unicodedata.category(c) != 'Mn'
    )

    name = re.sub(r"[-']", " ", name)

    name = re.sub(r"\s+", " ", name)
    return name.strip()

def normalize_saint(name: str) -> str:

    name = re.sub(r"\bst\b", "saint", name)
    name = re.sub(r"\bste\b", "saint", name)
    name = re.sub(r"\bsainte\b", "saint", name)
    return name

def city_key_no_space(name: str) -> str:
    n = normalize_saint(name)

    return n.replace(" ", "")


datasetA["city_key"] = datasetA["city_name_clean"]
datasetB["city_key"] = datasetB["Ville_clean"]

datasetA["city_key"] = datasetA["city_name_clean"].apply(city_key_no_space)
datasetB["city_key"] = datasetB["Ville_clean"].apply(city_key_no_space)

common_cities = set(datasetA["city_key"]) & set(datasetB["city_key"])
```

```python
dup_cities = datasetA.groupby("city_key")["city_code"].nunique()
dup_cities = dup_cities[dup_cities > 1].index.tolist()

print(f"One-to-many city count: {len(dup_cities)}")
print(dup_cities[:20]) 
```
